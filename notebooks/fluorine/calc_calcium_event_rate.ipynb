```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ret_df = pd.read_csv('./resources/recordings/RET/result.csv', header=[0], index_col=[0])
ret_df
```

```python
for cell_name, series in ret_df.iteritems():
    std = series.std(skipna=False)
    series /= std
    ret_df.loc[:, cell_name] = series

ret_df
```

```python
frames = ret_df.index
cell_ids = list(range(len(ret_df.columns)))


plt.figure(figsize=(8,8), dpi=200)
plt.pcolormesh(frames, cell_ids, ret_df.T, shading='auto', vmin=0.0, vmax=10.0)
plt.ylabel('cell id')
plt.xlabel('frames')

plt.title('Heatmap for RET')
plt.show()
```

```python
from oasis.functions import deconvolve
```

```python
ret_spikes = []
fc_spikes  = []
em_spikes  = []
```

```python
def rate_mean(rates):
    return sum(rates) / len(rates)
```

```python
for cell_name, series in ret_df.iteritems():
    _c, spikes, _b, _g, _lam = deconvolve(series.to_numpy(), g=(None, None), penalty=1)
    event = len(spikes[spikes > 0.01])
    rate = event / (len(spikes) / 20.0)
    ret_spikes.append(rate)

ret_spikes
rate_mean(ret_spikes)
```

```python
for cell_name, series in fc_df.iteritems():
    _c, spikes, _b, _g, _lam = deconvolve(series.to_numpy(), g=(None, None), penalty=1)
    event = len(spikes[spikes > 0.01])
    rate = event / (len(spikes) / 20.0)
    fc_spikes.append(rate)

fc_spikes
rate_mean(fc_spikes)
```

```python
for cell_name, series in em_df.iteritems():
    _c, spikes, _b, _g, _lam = deconvolve(series.to_numpy(), g=(None, None), penalty=1)
    event = len(spikes[spikes > 0.01])
    rate = event / (len(spikes) / 20.0)
    em_spikes.append(rate)

em_spikes
rate_mean(em_spikes)
```

```python
plt.figure(figsize=(4,4), dpi=200)
plt.plot(['FC', 'RET', 'EM'], [rate_mean(fc_spikes), rate_mean(ret_spikes), rate_mean(em_spikes)])
plt.title('Ca2+ Event Rate')
plt.ylabel('Ca2+ event rate (Hz)')
plt.show()
```
