```python
import os

def save_context_csv_file_path(animal_name, context_name):
    saved_dir = './resources/context_data/%s/' % animal_name
    os.makedirs(saved_dir, exist_ok=True)
    return '%s%s.csv' % (saved_dir, context_name)
```

```python
for animal_name in const.ANIMAL_NAMES:
    print('animal name: %s' % animal_name)
    
    file_path = './resources/%s_Longitudinal_Traces.csv' % animal_name
    csv = CaImageCSV(file_path)
    
    engrams = csv.engrams
    new_columns = pd.MultiIndex.from_tuples(engrams.items())
        
    for context_name in const.CONTEXTS:
        print('context name: %s' % context_name)
        
        context_data = csv.filtered_by_context(context_name)
        context_data.columns = new_columns
        csv_save_path = save_context_csv_file_path(animal_name, context_name)
        context_data.to_csv(csv_save_path)
        print('##### save csv to %s' % csv_save_path)
        
    print('################################################################################')
```
