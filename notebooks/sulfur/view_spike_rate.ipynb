```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file_path = './resources/spikes_data/ID181106CreA/ID181106CreA_A1postES.csv'
df = pd.read_csv(csv_file_path, header=[0], index_col=[0, 1], low_memory=False)
```

```python
def view_spike_rate(data, cell_count, title, threshold=0.05):
    values = np.array([])
    for _, row in data.iterrows():
        values = np.append(values, len(row[row > threshold]) / cell_count * 100)
        
    plt.plot(values, label=str(threshold))
```

```python
cell_count = len(df.columns)

plt.figure(figsize=(100, 15))

view_spike_rate(df, cell_count, title='spike rates per frame(> 0.05)', threshold=0.05)
view_spike_rate(df, cell_count, title='spike rates per frame(> 1.0)', threshold=1.0)

plt.legend()
plt.show()
```
