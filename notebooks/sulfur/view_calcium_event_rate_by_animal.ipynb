```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import utils.sulfur.constant as const
```

```python
all_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/all_event_rates_threshold005.csv', index_col=[0], header=[0])
engram_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/engram_event_rates_threshold005.csv', index_col=[0], header=[0])
non_engram_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/non_engram_event_rates_threshold005.csv', index_col=[0], header=[0])
```

```python
all_event_rate_df
```

```python
plt.figure(figsize=(25, 15))
for animal_name, rows in all_event_rate_df.iterrows():
    plt.plot(rows, label=animal_name)

plt.legend()
plt.grid()
# plt.show()
plt.savefig('./resources/event_rates/per_animal/all.png')
```

```python
for animal_name in const.ANIMAL_NAMES:
    plt.rcParams["font.size"] = 22
    plt.figure(figsize=(25, 10))
    
    engram_event_rates = engram_event_rate_df.loc[animal_name, :]
    non_engram_event_rates = non_engram_event_rate_df.loc[animal_name, :]
    
    plt.plot(engram_event_rates, marker='s', label='engram cells')
    plt.plot(non_engram_event_rates, marker='s', label='non-engram cells')
    plt.title(animal_name)
    plt.xticks(rotation=45)
    plt.legend()
    # plt.show()
    plt.savefig('./resources/event_rates/per_animal/%s.png' % animal_name)
```
