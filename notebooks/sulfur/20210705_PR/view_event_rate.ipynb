```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
all_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/all_event_rates_threshold005.csv', index_col=[0], header=[0])
engram_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/engram_event_rates_threshold005.csv', index_col=[0], header=[0])
non_engram_event_rate_df = pd.read_csv('./resources/event_rates/oasis_by_cells/non_engram_event_rates_threshold005.csv', index_col=[0], header=[0])
```

```python
all_event_rate_df
```

```python
print(len(all_event_rate_df.columns))
print(len(all_event_rate_df.mean()))
print(len(all_event_rate_df.sem()))
```

```python
x = all_event_rate_df.columns.to_numpy()
y = all_event_rate_df.mean().to_numpy()
error = all_event_rate_df.sem().to_numpy()

plt.rcParams["font.size"] = 22
plt.figure(figsize=(25, 10))
plt.errorbar(x, y, yerr=error, capsize=2)
plt.plot(x, y, marker="s", markersize=12, markeredgewidth=3, color='royalblue')
plt.ylim(0, 1.5)
plt.ylabel('Ca2+ Event Rate (Hz)')
plt.tight_layout()
plt.show()
```

```python
engram_x = engram_event_rate_df.columns
engram_y = engram_event_rate_df.mean()
engram_err = engram_event_rate_df.sem()

non_engram_x = non_engram_event_rate_df.columns
non_engram_y = non_engram_event_rate_df.mean()
non_engram_err = non_engram_event_rate_df.sem()

plt.rcParams["font.size"] = 22
plt.figure(figsize=(25, 10))

plt.errorbar(engram_x, engram_y, yerr=engram_err, capsize=2)
plt.plot(engram_x, engram_y, marker="s", markersize=12, markeredgewidth=3, label='engram cell', color='royalblue')

plt.errorbar(non_engram_x, non_engram_y, yerr=non_engram_err, capsize=2)
plt.plot(non_engram_x, non_engram_y, marker="s", markersize=12, markeredgewidth=3, label='non-engram cell', color='tomato')

plt.ylim(0, 1.5)
plt.ylabel('Calcium Event Rate (Hz)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
x = all_event_rate_df.columns.to_numpy()
y = all_event_rate_df.mean().to_numpy()
error = all_event_rate_df.sem().to_numpy()

engram_x = engram_event_rate_df.columns
engram_y = engram_event_rate_df.mean()
engram_err = engram_event_rate_df.sem()

non_engram_x = non_engram_event_rate_df.columns
non_engram_y = non_engram_event_rate_df.mean()
non_engram_err = non_engram_event_rate_df.sem()

fig, aix = plt.subplots(2, figsize=(25, 15), tight_layout=True, sharex='all')

aix[0].errorbar(x, y, yerr=error, capsize=2)
aix[0].plot(x, y, marker="s", markersize=12, markeredgewidth=3, color='royalblue')
aix[0].set_ylim(0, 1.5)
aix[0].set_ylabel('Ca2+ Event Rate (Hz)')
aix[0].set_title('Ca2+ event rates (threshold: 0.05)')

aix[1].errorbar(engram_x, engram_y, yerr=engram_err, capsize=2)
aix[1].plot(engram_x, engram_y, marker="s", markersize=12, markeredgewidth=3, label='engram cell', color='royalblue')

aix[1].errorbar(non_engram_x, non_engram_y, yerr=non_engram_err, capsize=2)
aix[1].plot(non_engram_x, non_engram_y, marker="s", markersize=12, markeredgewidth=3, label='non-engram cell', color='tomato')
aix[1].set_ylim(0, 1.5)
aix[1].set_ylabel('Ca2+ Event Rate (Hz)')
aix[1].set_title('Ca2+ event rates with engram (threshold: 0.05)')
aix[1].legend(loc='lower right')

plt.xticks(rotation=45)

# plt.show()
plt.savefig('./resources/event_rates/figure_threshold005.png')
```
