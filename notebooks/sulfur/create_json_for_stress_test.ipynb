```python
with open('./resources/video-compression-test-GREY.json') as f:
    default_config = json.load(f)
    
print(type(default_config))
```

```python
default_config['experimentName'] = 'miniscope_stress_test'
```

```python
default_config
```

```python
import pandas as pd
options = pd.read_csv('./resources/miniscope_stress_test_options.csv', header=0, index_col=None)
options
```

```python
import re

for index, row in options.iterrows():
    config = default_config.copy()
    
    record_length = row['record length']
    frame_rate = row['frame rate']
    frame_per_file = row['frame per file']
    compression = row['compression']
    name = 'ID%s_%s_%s_%s' % (record_length, frame_rate, frame_per_file, compression)
    config['animalName'] = name
    
    m = re.search(r'^\d+', record_length)
    record_length_i = int(m.group()) * 60
    
    frame_rate = frame_rate.replace('fps', 'FPS')
    
    config['recordLengthinSeconds'] = record_length_i
    config['devices']['miniscopes']['My V4 Miniscope']['framesPerFile'] = frame_per_file
    config['devices']['miniscopes']['My V4 Miniscope']['frameRate'] = frame_rate
    config['devices']['miniscopes']['My V4 Miniscope']['compression'] = compression
    
    with open('./resources/miniscope_config/%s.json' % name, 'w') as f:
        json.dump(config, f, indent=4)
```
