```python
import pandas as pd
import numpy as np
import os

from scipy.stats import f_oneway
from matplotlib import pyplot as plt
from matplotlib.ticker import PercentFormatter

from utils.CaImageCSV import CaImageCSV
from utils.CalculateEventRate import CalculateEventRate
import utils.sulfur.constant as const
```

```python
def create_context_hash():
    hash = {}
    for context_name in const.CONTEXTS:
        hash[context_name] = np.array([])
    return hash
```

```python
animal_names = const.ANIMAL_NAMES
event_rates = create_context_hash()
engram_event_rates = create_context_hash()
non_engram_event_rates = create_context_hash()
```

```python
for animal_name in animal_names:
    print('animal name: %s' % animal_name)
    origin_file_path = './resources/%s_Longitudinal_Traces.csv' % animal_name
    csv = CaImageCSV(origin_file_path)
    
    for context_name in const.CONTEXTS:
        calculator = CalculateEventRate(animal_name, context_name, csv.engrams)
        calculator.calc_event_rate()
        
        event_rates[context_name] = np.append(event_rates[context_name], calculator.event_rates_by_cells)
        engram_event_rates[context_name] = np.append(engram_event_rates[context_name], calculator.engram_event_rates_by_cells)
        non_engram_event_rates[context_name] = np.append(non_engram_event_rates[context_name], calculator.non_engram_event_rates_by_cells)
```

```python
def concat_event_rates_for_all_cell(event_rates):
    all_event_rates = np.array([])
    for rates in event_rates.values():
        all_event_rates = np.append(all_event_rates, rates)
    return all_event_rates
```

```python
def show_hist_for_event_rate(event_rates, title=None):
    fig, axes = plt.subplots(ncols=2, figsize=(15, 5))
    
    hist_range = (0, 4)
    bins = 8
    weights = np.ones(len(event_rates)) / len(event_rates)
    xlabel = 'Ca2+ event rates(Hz)'
    axes[0].hist(event_rates, range=hist_range, bins=bins, weights=weights, ec='navy')
    axes[1].hist(event_rates, range=hist_range, bins=bins, weights=weights, ec='navy', cumulative=True)
    
    axes[0].set_xlabel(xlabel)
    axes[1].set_xlabel(xlabel)
    
    if title is not None:
        axes[0].set_title(title)
        axes[1].set_title(title + ' (cumulative)')
    axes[0].yaxis.set_major_formatter(PercentFormatter(1))
    axes[1].yaxis.set_major_formatter(PercentFormatter(1))
    # plt.show()
    
    os.makedirs('./resources/event_rates/hist/', exist_ok=True)
    plt.savefig('./resources/event_rates/hist/%s' % title)
```

```python
all_event_rates = concat_event_rates_for_all_cell(event_rates)
all_engram_event_rates = concat_event_rates_for_all_cell(engram_event_rates)
all_non_engram_event_rates = concat_event_rates_for_all_cell(non_engram_event_rates)
```

```python
show_hist_for_event_rate(all_event_rates, 'all cells')
show_hist_for_event_rate(all_engram_event_rates, 'engram all cells')
show_hist_for_event_rate(all_non_engram_event_rates, 'non engram all cells')

for context_name in const.CONTEXTS:
    show_hist_for_event_rate(event_rates[context_name], title=('context: %s' % context_name)) 
```
